# Simple Neural Network


```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# distance in miles
distance = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

# Delivery times in minutes
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)

inputs = []
```

# Define a model

```python
# Define a model
model = nn.Sequential(nn.Linear(1,1))
```

# Loss function & optimization

```python
# define a loss function and optimizer
loss_function = nn.MSELoss()
optimizer = optim.SDG(model.parameters(), lr=0.001)
```

# Training Loop

```python
# Training loop
for epoch in range(500):
    # 0. reset optimizer
    optimizer.zero_grad()
    # 1. make prediction
    output = model(inputs)
    # 2. calculate loss - how bad was this prediction
    loss = loss_function(output, times)
    # 3. calculate adjustment (optimization)
    loss.backlward()
    # 4. update the model
    optimizer.step()
```

# Inference

```python
# during inference, we don't need grad.
with torch.no_grad():
    test_distance = torch.tensor([[25.0]], dtype=torch.float32)
    predicted_time = model(test_distance)
    print(f"Predicted time for 25 miles is {predicted_time.item():.1f} minutes")
```
